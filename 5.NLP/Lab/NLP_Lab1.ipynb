# B. DOCUMENT CLASSIFICATION WITH CALIFORNIA HOUSING DATASET

I used the [California Housing dataset](https://scikit-learn.org/stable/datasets/real_world.html#california-housing-dataset) to implement document classification in this lab

I. PREPARE THE DATA

```python
from sklearn.datasets import fetch_california_housing
```

```python
california_housing = fetch_california_housing()
data, target = california_housing.data, california_housing.target
```

```python
import numpy as np
labels = (target > np.median(target)).astype(int)
```

```python
data_str = [" ".join(map(str, row)) for row in data]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_str, labels, test_size=0.2, random_state=42)
```

II. Extract feature vectors

```python
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
count_vect = CountVectorizer()
X_train_count = count_vect.fit_transform(X_train)
tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_count)
X_train_tf = tf_transformer.transform(X_train_count)
X_train_tf.shape
```

III. Training model

```python
from sklearn import neighbors
clf = neighbors.KNeighborsClassifier()
clf.fit(X_train_tf, y_train)
```

IV. Prediction

```python
X_test_count = count_vect.transform(X_test)
X_test_tfidf = tf_transformer.transform(X_test_count)
X_test_tfidf.shape
```

```python
test_label = clf.predict(X_test_tfidf)
[(x, california_housing.data[x]) for x in test_label]
```

V. Evaluate

```python
accuracy = clf.score(X_test_tfidf, y_test)
print(f"Accuracy: {accuracy:.2f}")
```
